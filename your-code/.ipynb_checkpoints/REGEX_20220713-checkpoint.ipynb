```python
import re
```

```python
poem = "Once upon a midnight dreary, while I pondered, weak and weary," \
       "Over many a quaint and curious volume of forgotten lore—" \
       "While I nodded, nearly napping, suddenly there came a tapping," \
       "As of some one gently rapping, rapping at my chamber door." \
       "'Tis some visitor,' I muttered, 'tapping at my chamber door—" \
       "Only this and nothing more.'" \
       "" \
       "Ah, distinctly I remember it was in the bleak December;" \
       "And each separate dying ember wrought its ghost upon the floor." \
       "Eagerly I wished the mórrow;—vainly I had sought to borrow" \
       "From my books surcease of sorrow—sorrow for the lost Lenore—" \
       "For the rare and radiant maiden whom the angels name Lenore—" \
       "Nameless here for evermore."
```

```python
re.findall(r'[a-z]', poem) # \w
```

```python
re.findall(r'[a-zA-Z]', poem)
```

```python
re.findall(r'[A-Z].+?\b', poem) # '.' comodín
                                # .+ trae todos los caracteres (avaricioso)
                                # .+? trae todos los caracter pero de manera limitada (lazy)
                                # \b boundery
```

```python
re.split(r'[.,;-]', poem)
```

```python
pattern = r'.apping'
result = re.findall(pattern, poem)
result 
```

```python
pattern = r'napping|tapping' # "|"
result = re.findall(pattern, poem)
result
```

```python
# | OR ; ^ inicio de linea; $ final de linea

# \ escape character \. me busca literal el punto

# ? * + {} cuantificadores
```

```python
pattern = r'^Once'
result = re.findall(pattern, poem)
result
```

```python
pattern = r'^once'
result = re.findall(pattern, poem)
result
```

```python
text = """Carlos Perez Chuchu, 
        Diego Perez, 
        Gaby de la Cerda, 
        Carlos Velazquez Meow
        """
pattern = r'Carlos [A-Z].+?\b [A-Z].+?\b'
result = re.findall(pattern, text)
result
```

```python
text = "\
Mochila    1.051    48.25\
Playera     1.024   180.54\
Reloj%    0.524    47.20\
Gorra&    0.251    27.14\
Bufanda??    0.508    41.05\
Bufandita$$    1.348    00.05"
```

```python
pattern = r'[a-z]+'
result = re.findall(pattern, text, flags=re.IGNORECASE)
result
```

```python
pattern = r'\d+$'
result = re.findall(pattern, text)
result
```

```python
pattern = r'\d+\.\d+$'
result = re.findall(pattern, text)
result
```

```python
text_2 = re.sub(r'[$?%&]', '', text)
text_2
```

```python
re.findall(r'Buf[a-z]+', text)
```

```python
ejercicios = '''ejercicio1
ejercicio2
ejercicio
un_meme_chistoso
fotosdegatos.zip
ejerciciodiez
virus_cuidado
ejercicio22
otracosa
'''
```

```python
mates = 'Si sumamos dos número enteros... blablabla: (2+4)=6'

pattern = r'\(\d\+\d\)=\d'
result = re.findall(pattern, mates)
result
```

```python
mates = 'Si sumamos dos número enteros... blablabla: (200+400)=600'

pattern = r'\(\d+\+\d+\)=\d+'
result = re.findall(pattern, mates)
result
```

```python
mates = 'Si sumamos dos número enteros... blablabla: (200000+4000)=204000'

pattern = r'\(\d{1,6}\+\d{1,4}\)=\d{1,6}'
result = re.findall(pattern, mates)
result
```

```python
correos = 'user1@ironhack.com, user2@hotmail.com, users@yahoo.com, luis_ponce@gmail.com, chuchu12@meow.edu'

pattern = r'[\w]+@[\w]+\.[\w]+'
result = re.findall(pattern, correos)
result
```
